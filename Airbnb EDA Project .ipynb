# how has the number of reviews change over with time?

```python
df.columns
```

```python
# Ensure 'last review' is in datetime format
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

# Drop rows where last review is missing
df_review_time = df.dropna(subset=['last review'])

# Group by month and count reviews
monthly_reviews = df_review_time['last review'].dt.to_period('M').value_counts().sort_index()

# Convert PeriodIndex to datetime for plotting
monthly_reviews.index = monthly_reviews.index.to_timestamp()

# Plotting
plt.figure(figsize=(12, 6))
monthly_reviews.plot(kind='line', marker='o', color='teal')

plt.title("Number of Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Overviews:
    
# To understand review trends, we analyzed the last review date of listings over time.
# The line chart shows how frequently listings received reviews each month. 
# This helps identify seasonal trends, platform growth, or activity drops (e.g. during COVID-19 lockdowns). 
# The chart indicates peaks and declines in customer engagement over time.

```
